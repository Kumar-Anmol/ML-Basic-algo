```python
## Import libraries numpy, sklearn, matplotlib
import numpy as np
import sklearn
import matplotlib.pyplot as plt
```

```python
## generate data
data1=np.concatenate((np.linspace(-100,-51,50).reshape((50,1)),np.zeros((50,1))),axis=1)
data1=np.concatenate((data1,np.ones((50,1))),axis=1)
data2=np.concatenate((np.linspace(-49,49,99).reshape((99,1)),np.zeros((99,1))),axis=1)
data2=np.concatenate((data2,np.ones((99,1))*(-1)),axis=1)
data3=np.concatenate((np.linspace(51,100,50).reshape((50,1)),np.zeros((50,1))),axis=1)
data3=np.concatenate((data3,np.ones((50,1))),axis=1)
data=np.concatenate((data1,data2))
## data is the data matrix on which we will perform svm
data=np.concatenate((data,data3))
```

```python
## plot the data
## plot the data on x-axis and y-axis with (y=0))
newplt_ = plt.figure(figsize=(10, 10))
for i in range(199):
    if(data[i][2]==1):
        plt.plot(data[i][0],data[i][1], 'r.')
    else:
        plt.plot(data[i][0],data[i][1], 'g*')       
```

```python
## import SVC from sklearn.svm
from sklearn.svm import SVC

## create a model with kernel='rbf',gamma=0.0003, C = 400
## these gamma and C are the hyperparameters of the model
## and obtained by trial and error
model=sklearn.svm.SVC(kernel='rbf',gamma=0.0003, C = 400)
svc=model.fit(data[:,0:2],data[:,2])
```

```python
## training error
score=svc.score(data[:,0:2],data[:,2])
print("Training error is ",1-score)
```

```python
## support vectors
print("Support vectors are ",svc.support_vectors_)
```

```python
## plot the decision boundary
plt.plot(data[:,0:2],svc.decision_function(data[:,0:2]))
for i in range(199):
    if(data[i][2]==1):
        plt.plot(data[i][0],data[i][1], 'r.')
    else:
        plt.plot(data[i][0],data[i][1], 'g*') 
plt.title("SVM classifier with RBF kernel (gamma=0.0003, C=400)")
plt.xlabel("X")
plt.ylabel("Y")
plt.show() 
```
